```python
import numpy as np

def plot_stability(file_name):
    responses = np.load(f"test_cache/permute/{file_name}.npz")
    train_averaged = np.average(responses["train"], axis = 0)
    init_averaged = np.average(responses["init"], axis = 0)

    train_std = np.std(responses["train"], axis = 0)
    init_std = np.std(responses["init"], axis = 0)

    import matplotlib.pyplot as plt
    plt.plot(train_averaged, color = "blue")
    plt.fill_between(range(len(train_averaged)), np.clip(train_averaged - train_std, 0, 1), np.clip(train_averaged + train_std, 0, 1),
                    color='blue', alpha=0.1)
    plt.plot(init_averaged, color = "red")
    plt.fill_between(range(len(init_averaged)), np.clip(init_averaged - init_std, 0, 1), np.clip(init_averaged + init_std, 0, 1),
                    color='red', alpha=0.1)
    plt.xlabel("Edits")
    plt.ylabel("Recall Success (over N sequences)")
    plt.title(f"File - {file_name}")
```

```python
plot_stability("permute_end deplete_200_5")
```

```python
plot_stability("permute_end deplete_200_50")
```
